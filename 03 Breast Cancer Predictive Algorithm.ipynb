Comparison of gradient descent and normalized gradient descent using the softmax cost function with a real breast cancer dataset

```python
import basic_gradient_descent
import autograd.numpy as np   
import matplotlib.pyplot as plt
```

```python
csvname = 'breast_cancer_data.csv'
data = np.loadtxt(csvname,delimiter = ',')

x = data[:,:-1].T
y = data[:,-1:] 

bgd = basic_gradient_descent.Setup(x,y)
```

```python
g = bgd.softmax; w = 0.1*np.random.randn(x.shape[0]+1,1); max_its = 100; alpha_choice = 10**(-1);
weight_history_1,cost_history_1 = bgd.gradient_descent(g,alpha_choice,max_its,w)
count_history_1 = [bgd.counting_cost(v) for v in weight_history_1]
```

```python
g = bgd.softmax; alpha_choice = 1;
weight_history_2,cost_history_2,xn = bgd.normalized_gradient_descent(g,alpha_choice,max_its,w)
count_history_2 = [bgd.counting_cost(v) for v in weight_history_2]
```

```python
vals = np.linspace(0,100,len(cost_history_1))
fig = plt.figure(figsize = (7,3))
plt.plot(vals,cost_history_1)
plt.plot(vals,cost_history_2)
plt.legend(['standard' , 'normalized'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('cost function history')
plt.show()
```

```python
vals = np.linspace(0,100,len(count_history_1))
fig = plt.figure(figsize = (7,3))
plt.plot(vals,count_history_1)
plt.plot(vals,count_history_2)
plt.legend(['standard' , 'normalized'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('misclassification history')
plt.show()
```
