Basic gradient descent algorithm to minimize a least squares cost function for a linear regression. Toy data used as an example.

```python
import basic_gradient_descent
import autograd.numpy as np
import matplotlib.pyplot as plt
```

```python
csvname = '2d_linregress_data.csv'
data = np.loadtxt(csvname,delimiter = ',')

# get input and output of dataset
x = data[:,:-1].T
y = data[:,-1:] 

bgd = basic_gradient_descent.Setup(x,y)

# scatter plot the input data
plt.figure()
plt.scatter(x,y,color = 'k',edgecolor = 'w')
plt.show()
```

```python
g = bgd.least_squares; max_its = 1000; a = 10**-0.5; w = 0.1*np.random.randn(2,1);
weight_history,cost_history = bgd.gradient_descent(g,a,max_its,w)
```

```python
its = np.linspace(0,1000,len(cost_history))
fig = plt.figure(figsize = (7,3))
plt.plot(its,cost_history)
plt.legend(['a = 10^-0.5'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
print("the coefficient vector minimizing the least squares cost function is given by w = ", [weight_history[-1][0][0], weight_history[-1][1][0]])
plt.figure()
xt = np.linspace(0,1)
yt = weight_history[-1][0] + weight_history[-1][1] * xt

plt.scatter(x,y,color = 'k',edgecolor = 'w')
plt.plot(xt,yt)
plt.legend(['best fit'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

```
