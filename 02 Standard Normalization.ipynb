Basic standard normalization technique utilized to improve minimized least squares value via gradient descent. Toy student debt data is used in this example.

```python
import basic_gradient_descent
import autograd.numpy as np
import matplotlib.pyplot as plt
```

```python
csvname = 'student_debt.csv'
data = np.loadtxt(csvname,delimiter = ',')

x = data[:,:-1].T
y = data[:,-1:] 

bgd = basic_gradient_descent.Setup(x,y)

plt.figure()
plt.scatter(x,y,color = 'k',edgecolor = 'w')
plt.show()
```

```python
g = bgd.least_squares; max_its = 100; a = 10**-7; w = np.array([0.0,0.0])[:,np.newaxis];
weight_history,cost_history = bgd.gradient_descent(g,a,max_its,w)
```

```python
vals = np.linspace(0,100,len(cost_history))
fig = plt.figure(figsize = (7,3))
plt.plot(vals,cost_history)
plt.legend(['a = 10^-7'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
plt.figure()
yt1 = weight_history[-1][0] + weight_history[-1][1] * x[0]
plt.scatter(x,y,color = 'k',edgecolor = 'w')
plt.plot(x[0],yt1)
plt.legend(['nonstandardized prediction'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
g = bgd.least_squares; w = np.array([0.0,0.0])[:,np.newaxis]; max_its = 100; alpha_choice = 10**(-1);
weight_history,cost_history, x_normalized = bgd.normalized_gradient_descent(g,alpha_choice,max_its,w)
```

```python
plt.figure()
yt2 = weight_history[-1][0] + weight_history[-1][1] * x_normalized[0]

plt.plot(x[0], yt1)
plt.plot(x[0], yt2)

plt.scatter(x,y,color = 'k',edgecolor = 'w')
plt.legend(['nonstandardized input' , 'standardized input'],loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
